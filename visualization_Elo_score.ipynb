```python
# Load Elo Score dataset
data = pd.read_csv('elo.csv')
```

```python
names_list = list(pd.unique(data['Home/Neutral']))
    
def extract_elo(row, team_name):
    if row['Home/Neutral'] == team_name:
        return row['elo_home']
    elif row['Visitor/Neutral'] == team_name:
        return row['elo_away']
    else:
        return 0

elo_team = []   
 
for name in names_list:
    elo_team.append(data.apply(extract_elo, team_name = name, axis = 1))

for i in range(len(elo_team)):
    elo_team[i] = elo_team[i][elo_team[i] > 0]
    elo_team[i] = elo_team[i].reset_index(drop = True)
    
```
