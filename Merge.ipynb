```python
import glob, os
import pandas as pd
import numpy as np

df1 = pd.read_csv('Project Data/Game_results13-14.csv', parse_dates=True)
df1['Season'] = '2013/2014'
df2 = pd.read_csv('Project Data/Game_results14-15.csv', parse_dates=True)
df2['Season'] = '2014/2015'
df3 = pd.read_csv('Project Data/Game_results15-16.csv', parse_dates=True)
df3['Season'] = '2015/2016'
df4 = pd.read_csv('Project Data/Game_results16-17.csv', parse_dates=True)
df4['Season'] = '2016/2017'
df5 = pd.read_csv('Project Data/Game_results17-18.csv', parse_dates=True)
df5['Season'] = '2017/2018'
df = pd.concat([df1, df2, df3, df4, df5], ignore_index=True)
df['Unnamed: 7'] = df['Unnamed: 7'].apply(lambda x: 1 if x == 'OT' else (2 if x == '2OT' else (3 if x == '3OT'
                                          else (4 if x == '4OT' else 0))))
df['Visitor/Neutral'] = df['Visitor/Neutral'].apply(lambda x: 'Charlotte Hornets' if x == 'Charlotte Bobcats' else x)
df['Home/Neutral'] = df['Home/Neutral'].apply(lambda x: 'Charlotte Hornets' if x == 'Charlotte Bobcats' else x)
df.rename(columns={'Unnamed: 7':'Overtime', 'PTS':'Visitor-Points', 'PTS.1':'Home-Points'}, inplace=True)
df = df.drop(['Unnamed: 6', 'Notes'], 1)
df.head()
```

```python
df1M = pd.read_csv('Project Data/M_table13-14.csv')
df1M.columns = df1M.iloc[0]
df1M['Season']='2013/2014'
df2M = pd.read_csv('Project Data/M_table14-15.csv')
df2M.columns = df2M.iloc[0]
df2M['Season']='2014/2015'
df3M = pd.read_csv('Project Data/M_table15-16.csv')
df3M.columns = df3M.iloc[0]
df3M['Season']='2015/2016'
df4M = pd.read_csv('Project Data/M_table16-17.csv')
df4M.columns = df4M.iloc[0]
df4M['Season']='2016/2017'
df5M = pd.read_csv('Project Data/M_table17-18.csv')
df5M.columns = df5M.iloc[0]
df5M['Season']='2017/2018'
dfM = pd.concat([df1M, df2M, df3M, df4M, df5M], ignore_index=True)
dfM = dfM[dfM.Team != 'Team']
dfM = dfM[dfM.Team != 'League Average']
dfM.Team = dfM.Team.str.replace('*','')
dfM['Team'] = dfM['Team'].apply(lambda x: 'Charlotte Hornets' if x == 'Charlotte Bobcats' else x)
dfM = dfM.drop(['Rk', 'Arena', 'Attend.', 'Attend./G'], 1)
dfM.columns.values[15] = "OeFG%"
dfM.columns.values[16] = "OTOV%"
dfM.columns.values[17] = "ODRB%"
dfM.columns.values[18] = "OFT/FGA"
dfM.columns.values[19] = "DeFG%"
dfM.columns.values[20] = "DTOV%"
dfM.columns.values[21] = "DDRB%"
dfM.columns.values[22] = "DFT/FGA"
dfM.head()
```

```python
df1O = pd.read_csv('Project Data/O_table13-14.csv')
df1O['Season']='2013/2014'
df2O = pd.read_csv('Project Data/O_table14-15.csv')
df2O['Season']='2014/2015'
df3O = pd.read_csv('Project Data/O_table15-16.csv')
df3O['Season']='2015/2016'
df4O = pd.read_csv('Project Data/O_table16-17.csv')
df4O['Season']='2016/2017'
df5O = pd.read_csv('Project Data/O_table17-18.csv')
df5O['Season']='2017/2018'
dfO = pd.concat([df1O, df2O, df3O, df4O, df5O], ignore_index=True)
dfO.Team = dfO.Team.str.replace('*','')
dfO = dfO[dfO.Team != 'League Average']
dfO['Team'] = dfO['Team'].apply(lambda x: 'Charlotte Hornets' if x == 'Charlotte Bobcats' else x)
dfO = dfO.drop(['Rk'], 1)
dfO = dfO.add_prefix('O_')
dfO.head()
```

```python
df1T = pd.read_csv('Project Data/T_table13-14.csv')
df1T['Season']='2013/2014'
df2T = pd.read_csv('Project Data/T_table14-15.csv')
df2T['Season']='2014/2015'
df3T = pd.read_csv('Project Data/T_table15-16.csv')
df3T['Season']='2015/2016'
df4T = pd.read_csv('Project Data/T_table16-17.csv')
df4T['Season']='2016/2017'
df5T = pd.read_csv('Project Data/T_table17-18.csv')
df5T['Season']='2017/2018'
dfT = pd.concat([df1T, df2T, df3T, df4T, df5T], ignore_index=True)
dfT.Team = dfT.Team.str.replace('*','')
dfT = dfT[dfT.Team != 'League Average']
dfT['Team'] = dfT['Team'].apply(lambda x: 'Charlotte Hornets' if x == 'Charlotte Bobcats' else x)
dfT = dfT.drop(['Rk'], 1)
dfT = dfT.add_prefix('T_')
dfT.head()
```

```python
df_new = pd.merge(df, dfM, how='left', left_on=['Visitor/Neutral','Season'], right_on = ['Team','Season'])
df_new = pd.merge(df_new, dfM, how='left', left_on=['Home/Neutral','Season'], right_on = ['Team','Season'])
df_new = df_new.drop(['Team_x', 'Team_y'], 1)
df_new.head()
```

```python
df_new = pd.merge(df_new, dfO, how='left', left_on=['Visitor/Neutral','Season'], right_on = ['O_Team','O_Season'])
df_new = pd.merge(df_new, dfO, how='left', left_on=['Home/Neutral','Season'], right_on = ['O_Team','O_Season'])
df_new = pd.merge(df_new, dfT, how='left', left_on=['Visitor/Neutral','Season'], right_on = ['T_Team','T_Season'])
df_new = pd.merge(df_new, dfT, how='left', left_on=['Home/Neutral','Season'], right_on = ['T_Team','T_Season'])
df_new = df_new.drop(['O_Team_x', 'O_Team_y', 'T_Team_x', 'T_Team_y',
                      'O_Season_x', 'O_Season_y', 'T_Season_x', 'T_Season_y'], 1)
df_new.columns = [c.replace('_x', '_Visitor') for c in df_new.columns]
df_new.columns = [c.replace('_y', '_Home') for c in df_new.columns]
df_new.head()
```

```python
df_new.to_csv('example.csv', index=False)
```
